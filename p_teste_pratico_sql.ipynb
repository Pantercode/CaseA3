# FUN√á√ÉO PARA INSERIR REGISTROS NO BANCO DE DADOS

```python
# Recebe como par√¢metro
# 1. Nome do banco de dados
# 2. Nome da tabela que receber√° os dados
# 3. Lista das colunas da tabela
# 4. Dados que ser√£o inseridos

def baixar_dados(nome_arquivo, url):
    resposta = requests.get(url)
    with open(nome_arquivo, 'wb') as arquivo:
        arquivo.write(resposta.content)

```

# FUN√á√ÉO PARA BAIXAR A BASE DE DADOS UTILIZADA PARA DESENVOLVIMENTO DAS ATIVIDADES

```python
# Recebe como par√¢metro:
# 1. Nome do arquivo com a extens√£o (Ex: base.csv)
# 2. URL do site onde o dado est√° localizado (Ex: 'https://drive.google.com/user_teste/folder_dataset/file.csv'):

def inserir_registros(banco, tabela, lista_colunas, arquivo):
    conexao_sql = sqlite3.connect(banco)
    cursor = conexao_sql.cursor()
    with open(arquivo, 'r') as file:
        dados = csv.reader(file, delimiter=';')  # Especificar o delimitador como ponto e v√≠rgula
        next(dados)  # Pular o cabe√ßalho se houver
        conteudo_dados = [tuple(linha) for linha in dados]  # Transformar cada linha em uma tupla
        # Imprimir algumas linhas para diagn√≥stico
        print("Algumas linhas de exemplo:", conteudo_dados[:5])

    num_colunas = lista_colunas.count(',') + 1  # Ajustar a contagem de colunas
    string_insert = 'INSERT INTO ' + tabela + ' (' + lista_colunas + ') VALUES (' + ', '.join(['?'] * num_colunas) + ')'
    print("String SQL:", string_insert)  # Imprimir a string SQL para diagn√≥stico

    cursor.executemany(string_insert, conteudo_dados)
    conexao_sql.commit()
    conexao_sql.close()
```

## Pergunta 1:


 Escreva um comando em SQL que retorne a quantidade de vendas dos estados de Minas Gerais e S√£o Paulo por ano e m√™s

```python
consulta = '''
SELECT
  strftime('%Y', v.data_compra) AS Ano,
  strftime('%m', v.data_compra) AS Mes,
  u.estado,
  SUM(v.quantidade) AS Quantidade_Total
FROM vendas v
JOIN usuarios u ON v.cod_usuario = u.cod_usuario
WHERE u.estado IN ('Minas Gerais', 'S√£o Paulo')
GROUP BY
  strftime('%Y', v.data_compra),
  strftime('%m', v.data_compra),
  u.estado
ORDER BY
  Ano,
  Mes,
  u.estado;
'''

cursor.execute(consulta)



df = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])
df.head()
```

## Pergunta 4:

 Escreva um comando em SQL que retorne o ticket m√©dio (valor m√©dio das compras) por faixa et√°ria.

```python

consulta = '''
WITH FaixaEtaria AS (
  SELECT
    v.cod_usuario,
    v.valor,
    CASE
      WHEN u.faixa_etaria = 'Entre 18 a 25 anos' THEN '18-25'
      WHEN u.faixa_etaria = 'Entre 26 a 30 anos' THEN '26-30'
      WHEN u.faixa_etaria = 'Entre 31 a 35 anos' THEN '31-35'
      WHEN u.faixa_etaria = 'Entre 36 a 45 anos' THEN '36-45'
      WHEN u.faixa_etaria = 'Entre 46 a 55 anos' THEN '46-55'
      WHEN u.faixa_etaria = 'Entre 56 a 65 anos' THEN '56-65'
      WHEN u.faixa_etaria = 'Entre 61 a 70 anos' THEN '61-70'
      ELSE '65+'
    END AS faixa
  FROM vendas v
  JOIN usuarios u ON v.cod_usuario = u.cod_usuario
)

SELECT
  faixa,
  ROUND(AVG(valor),2) AS ticket_medio
FROM FaixaEtaria
GROUP BY faixa
ORDER BY faixa;


'''
cursor.execute(consulta)

# Carregar os dados do cursor em um DataFrame do pandas
df = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])
df.head()
```

## Pergunta 5:


Escreva um comando em SQL que retorna o produto, quantos usu√°rios fizeram pelo menos uma compra desse produto e qual √© o valor m√≠nimo e m√°ximo pago

```python

consulta = '''

SELECT
  p.nome_produto,
  COUNT(DISTINCT v.cod_usuario) AS numero_usuarios,
  ROUND(MIN(v.valor),2) AS valor_minimo,
  ROUND(MAX(v.valor),2) AS valor_maximo
FROM vendas v
JOIN produtos p ON v.cod_produto = p.cod_produto
GROUP BY p.nome_produto
ORDER BY numero_usuarios DESC;


'''
cursor.execute(consulta)

# Carregar os dados do cursor em um DataFrame do pandas
df = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])
df.head()
```

## Pergunta 7:

Escreva um comando em SQL que retorne quais s√£o os 3 produtos mais comprados dos estados da regi√£o Sul e Sudeste, a quantidade de itens comprados, o valor total pago e a m√©dia de pre√ßo paga

```python

consulta = '''

SELECT
  v.cod_produto,
  p.nome_produto,
  ROUND(SUM(v.quantidade),2) AS quantidade_total,
  ROUND(SUM(v.valor),2) AS valor_total,
  ROUND(AVG(v.valor / v.quantidade),2) AS preco_medio
FROM vendas v
JOIN usuarios u ON v.cod_usuario = u.cod_usuario
JOIN produtos p ON v.cod_produto = p.cod_produto
WHERE u.estado IN ('Rio Grande do Sul', 'Santa Catarina', 'Paran√°', 'S√£o Paulo', 'Rio de Janeiro', 'Minas Gerais', 'Esp√≠rito Santo')
GROUP BY v.cod_produto, p.nome_produto
ORDER BY quantidade_total DESC
LIMIT 3;


'''
cursor.execute(consulta)

# Carregar os dados do cursor em um DataFrame do pandas
df = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])
df.head()
```

## Pergunta 8

Escreva um comando em SQL que retorne o produto, a quantidade de itens vendidos e a receita do produto que tem mais itens vendidos por ano e m√™s

```python
consulta = '''

WITH Meses_bons AS (
    SELECT
        strftime('%Y', data_compra) AS Ano,
        strftime('%m', data_compra) AS Mes,
        cod_produto,
        SUM(quantidade) AS quantidade_total,
        ROUND(SUM(valor),2)AS receita_total
    FROM vendas
    GROUP BY Ano, Mes, cod_produto
),

Ranking AS (
    SELECT *,
        (SELECT COUNT(*)
         FROM Meses_bons mb2
         WHERE mb2.Ano = mb1.Ano AND mb2.Mes = mb1.Mes AND
               (mb2.quantidade_total > mb1.quantidade_total OR
                (mb2.quantidade_total = mb1.quantidade_total AND mb2.cod_produto < mb1.cod_produto))
        ) + 1 AS rank
    FROM Meses_bons mb1
)

SELECT
    r.Ano,
    r.Mes,
    p.nome_produto,
    r.quantidade_total,
    r.receita_total
FROM Ranking r
JOIN produtos p ON r.cod_produto = p.cod_produto
WHERE r.rank = 1;



'''
cursor.execute(consulta)

# Carregar os dados do cursor em um DataFrame do pandas
df = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])
df.head()
```
